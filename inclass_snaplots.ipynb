```python
df=pd.read_csv('diabetes.csv')
```

```python
for i in df.columns:
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  outliers=(df[i]<( Q1 - 1.5*IQR)) | (df[i]>(Q3+1.5*IQR))
  print(i,outliers.sum())
```

```python
df1=df.copy()
print(df1.shape)
for i in df.columns:
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  outliers=(df1[i]<( Q1 - 1.5*IQR)) | (df1[i]>(Q3+1.5*IQR))
  print(i,outliers.sum())
  df1=df1.drop(df1[outliers].index)
  print(df1.shape)
```

This next plot shows all the different possible pairs of scatter plots using all of the different columns, excluding the 'outcome' Column.

```python
# plt.Figure(figsize=(15,10))
sns.pairplot(data=df,
             vars = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"],  # names of columns used for the plot
             kind="scatter",                        # kind of plots for each pair of different columns
             diag_kind="kde",                       # kind of plots on the diagonal
             #hue="sex",
             height=2,
             aspect=1.1,
             palette="muted"
             )

plt.suptitle("Pair plot", y=1.02)   # adds title to the plot
plt.show()
```

This next plot is a bar plot showing the mean values of each column.

```python
df2 = df.copy()
means = df2.mean().tolist()

plt.Figure(figsize=(15, 10))
sns.barplot(x=df2.columns, y=means)
plt.xlabel('Columns')
plt.ylabel('Mean Values')
plt.title('Mean of Each Column')
plt.xticks(rotation=45)
plt.show()
```

This next plot shows how all the other factors: Such as pregnancy, BloodPressure, Insulin, BMI, DiabetesPedigreeFunction, and Age react depending on the different insulin levels.

```python
columns_to_plot = ['Pregnancies', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age']
plt.figure(figsize=(15, 10))

for column in columns_to_plot:
    sns.lineplot(data=df2, x='Insulin', y=column, label=column)

plt.xlabel('Insulin')
plt.ylabel('Values')
plt.title('Multiple Line Plots with Insulin as X-Axis')
plt.legend(title='Features')
plt.show()
```

This next plot shows a violin plot, using the df1 data which is without outliers. The violin plot is comparable to the box plot, but shows the probability distribution as well.

```python
plt.Figure(figsize=(15,10))
sns.violinplot(data=df1)
plt.xticks(rotation=45)
plt.show()
```

This last plot is a FacetGrid, where we are going to bin the BMI into 5 categories of very low, low, medium, high, and very high. This way it turns 'Categorical'. And we'll use outcome as well. And using histograms for Insulin in this context to visualize the distribution of Insulin levels across different categories of BMI and Outcome


```python
plt.figure(figsize=(15, 10))

# Bin the BMI into categories
df['BMI_binned'] = pd.cut(df['BMI'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

plt.figure(figsize=(15, 10))
g = sns.FacetGrid(data=df, col='BMI_binned', row='Outcome', margin_titles=True)
g.map(sns.histplot, 'Insulin')

plt.show()
```
